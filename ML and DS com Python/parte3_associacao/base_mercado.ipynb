```python
import pandas as pd
from apyori import apriori


base_mercado1 = pd.read_csv('mercado.csv')
print(base_mercado1)
```

```python
transacoes = []
for i in range(len(base_mercado1)):
    transacoes.append([str(base_mercado1.values[i, j]) for j in range(base_mercado1.shape[1])])
    
print(transacoes)
```

```python
regras = apriori(transacoes, min_support=0.3, min_confidence=0.8, min_lift=2)
resultados = list(regras)
print(resultados)
```

```python
len(resultados)
```

```python
resultados[0]
```

```python
A = []
B = []
suporte = []
confianca = []
lift = []

for resultado in resultados:
    s = resultado[1]
    result_rules = resultado[2]
    for result_rule in result_rules:
        a = list(result_rule[0])
        b = list(result_rule[1])
        c = result_rule[2]
        l = result_rule[3]
        A.append(a)
        B.append(b)
        confianca.append(c)
        suporte.append(s)
        lift.append(l)
```

```python
rules_df = pd.DataFrame({'A': A, 'B': B,  'suporte': suporte, 'confianca': confianca, 'lift': lift})

rules_df
```

```python
A = []
B = []
suporte = []
confianca = []
lift = []

for resultado in resultados:
    s = resultado[1]
    result_rules = resultado[2]
    for result_rule in result_rules:
        a = list(result_rule[0])
        b = list(result_rule[1])
        c = result_rule[2]
        l = result_rule[3]
        A.append(a)
        B.append(b)
        confianca.append(c)
        suporte.append(s)
        lift.append(l)
```
